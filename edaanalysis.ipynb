```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/dataset.csv"
df = pd.read_csv(file_path)

# Summary statistics
print(df.describe())

# Plot distributions
df.hist(figsize=(20, 11), bins=20, edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["AGE"], y=df["SMOKING"], hue=df["LUNG_CANCER"])
plt.title("Scatter Plot: Age vs Smoking")
plt.show()

```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x="GENDER", hue="LUNG_CANCER", data=df)
plt.title("Bar Chart: Lung Cancer Count by Gender")
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x="SMOKING", hue="LUNG_CANCER", data=df, palette="coolwarm")

plt.xlabel("Smoking Status (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Lung Cancer Cases by Smoking Status")
plt.legend(title="Lung Cancer")
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x="ALCOHOL CONSUMING", hue="LUNG_CANCER", data=df, palette="viridis")

plt.xlabel("Alcohol Consuming (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Lung Cancer Cases by Alcohol Consumption")
plt.legend(title="Lung Cancer")
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x="WHEEZING", hue="LUNG_CANCER", data=df, palette="magma")

plt.xlabel("Wheezing (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Lung Cancer Cases by Wheezing")
plt.legend(title="Lung Cancer")
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x="CHEST PAIN", hue="LUNG_CANCER", data=df, palette="rocket")

plt.xlabel("Chest Pain Level")
plt.ylabel("Count")
plt.title("Lung Cancer Cases by Chest Pain")
plt.legend(title="Lung Cancer")
plt.show()



```

```python
plt.figure(figsize=(6, 6))
df["LUNG_CANCER"].value_counts().plot.pie(autopct="%1.1f%%", colors=["skyblue", "salmon"])
plt.title("Lung Cancer Distribution")
plt.ylabel("")  # Hide Y-label
plt.show()
```

```python
sns.pairplot(df, hue="LUNG_CANCER", diag_kind="kde")
plt.show()

```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x="LUNG_CANCER", y="AGE", data=df, palette="pastel")

plt.xlabel("Lung Cancer (NO = 0, YES = 1)")
plt.ylabel("Age")
plt.title("Violin Plot: Age Distribution by Lung Cancer Status")
plt.show()
```
